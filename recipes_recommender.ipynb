
# Weighted Rating(WR)=(v/(v+m) * R) + (m/(v+m) * C) where:
- v is the number of voters for the book
- m is the minimum votes required to be listed in the chart
- R is the average rating of the book
- C is the mean vote across the whole report

```python
rec=recipes.groupby('title').mean()
rec['votes']=recipes.groupby('title').count()['calories']
rec=rec.reset_index()
rec.head()
```

```python
C = rec['rating'].mean()
C
```

```python
m = rec['votes'].quantile(0.95)
m
```

```python
q_rec = rec.copy().loc[rec['votes'] >= m]
q_rec.shape
```

```python
def weighted_rating(x, m=m, C=C, coeffs=[0.1,0.1,0.7,0.1]):
    v = x['votes']
    R = x['rating']
    
    wr = (v/(v+m) * R) + (m/(v+m) * C)
    return wr * (x['calories']*coeffs[0] * x['fat'] * coeffs[1] * x['protein'] * coeffs[2] * x['sodium']* coeffs[3])
```

```python
q_rec['score'] = q_rec.apply(weighted_rating, axis=1)
recipes[recipes['title'].isin(list(q_rec.sort_values('score',ascending=False).head(15)['title']))]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
recipes['title'] = recipes['title'].fillna('')
tfidf_matrix = tfidf.fit_transform(recipes['title'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(recipes.index, index=recipes['title']).drop_duplicates()
indices
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:10]
    recipes_indices = [i[0] for i in sim_scores]
    return recipes.iloc[recipes_indices]
```

# Make string from ingredients and categories columns

```python
from nltk.corpus import wordnet as wn
#food = wn.synset('food.n.02')
food = wn.synset('food.n.01')
foodlist1=list(set([w.lower() for s in wn.synset('food.n.01').closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))
foodlist2=list(set([w.lower() for s in wn.synset('food.n.02').closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))
foodlist3=list(set([w.lower() for s in wn.synset('food.n.03').closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))
food = sum([foodlist1, foodlist2, foodlist3], [])
food
```

```python
import string
def getfood(x):
    ingredients=x['ingredients']
    lista=[]
    for row in ingredients:
        row=row.lower().translate(str.maketrans('', '', string.punctuation))
        for el in row.split():
            if el in food:
                lista.append(el)
    return ' '.join(list(set(lista)))
    
def getcategories(x):
    categories=x['categories']
    clista=[]
    for c in categories:
        clista.append(c.lower().replace(" ",""))
        
    return ' '.join(list(set(clista)))
    
recipes['ingredients_list']=recipes.apply(getfood,axis=1)
recipes['categories_list']=recipes.apply(getcategories,axis=1)

recipes.head()
```

# Recommendations based on KNN classifier

```python
temp=recipes.groupby('title')['categories_list'].apply(lambda x: ' '.join(x)).reset_index()
names=temp['title'].tolist()
```

```python
brojac = CountVectorizer(stop_words='english')
count_matrix = brojac.fit_transform(temp['categories_list'])
```

```python
values=count_matrix.toarray()
#values[np.where(values!=0)]=1
```

```python
empty=pd.DataFrame(data=values,columns=brojac.get_feature_names(),index=names)
empty.shape
```

```python
empty.head()
```

```python
from scipy.sparse import csr_matrix
data_matrix = csr_matrix(empty.values)
data_matrix.toarray().max()
```

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine')#, algorithm = 'brute')
model_knn.fit(data_matrix)
```

```python
import numpy as np
query_index = np.random.choice(empty.shape[0])
distances, indices = model_knn.kneighbors(empty.iloc[query_index, :].values.reshape(1, -1), n_neighbors = 17774)

top=10
top_counter=0
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(empty.index[query_index]))
    else:
        if top_counter>=top:
            break
        set1=set(str(temp[temp['title']==empty.index[query_index]]['categories_list'].values)[2:-2].split(" "))
        set2=set(str(temp[temp['title']==empty.index[i]]['categories_list'].values)[2:-2].split(" "))
        if len(set1.intersection(set2))>=3:
            print('{0}: {1}, with distance of {2}:'.format(i, empty.index[indices.flatten()[i]], distances.flatten()[i]))
            print('Food {0} and food {1} have in common those cathegories:'.format(empty.index[query_index],empty.index[i]))
            for e,j in enumerate(set1.intersection(set2)):
                print('\t {}.{} {}'.format(i,e,j))
            print('Food {0} dont have in common those categories:'.format(empty.index[query_index]))
            for e,j in enumerate(set2.difference(set1)):
                print('\t {}.{} {}'.format(i,e,j))
            print('\n')
            top_counter+=1
```
